## ***Introduction du Projet : Débruitage d'Images avec un Autoencodeur Convolutif***

## **Résumé des Étapes Réalisées dans le Projet : Classification Automatique d'Images**
Ce projet vise à développer une solution efficace pour la classification d'images en cinq catégories : chats, chiens, portraits, voitures, et fleurs, en utilisant des réseaux neuronaux convolutifs (CNN). Voici un résumé détaillé des étapes réalisées :

**1. Préparation des Données**

- Les images sont chargées depuis plusieurs dossiers représentant différentes catégories (chats, chiens, portraits, voitures, fleurs). Chaque catégorie - est associée à un label unique.
- Les images sont redimensionnées à une taille uniforme de 128x128 pixels, et normalisées pour que leurs valeurs de pixels soient comprises entre 0 et 1.
- Les images sont ensuite séparées en ensembles d'entraînement (80%) et de validation (20%) à l'aide de la fonction train_test_split de scikit-learn.
- Un ensemble de test supplémentaire est préparé avec des images provenant d'un dossier spécifique.

**2. Augmentation des Données d'Entraînement**

- Un générateur de données d'augmentation (ImageDataGenerator) est utilisé pour créer des variations des images d'entraînement via des transformations telles que des rotations, des déplacements horizontaux et verticaux, des changements de zoom, et des retournements horizontaux.
- Cela permet d'augmenter la diversité des données d'entraînement, réduisant ainsi le risque de surapprentissage.

***3. Définition du Modèle d'Autoencodeur***

- Un autoencodeur convolutif est construit en deux parties : un encodeur et un décodeur.
    - Encodeur : Comprend des couches convolutives et de pooling pour extraire les caractéristiques des images.
    - Décodeur : Reconstruit les images à partir des caractéristiques extraites, en utilisant des couches convolutives et de sur-échantillonnage.
- L'autoencodeur est compilé avec l'optimiseur 'adam' et la fonction de perte 'mse' (Mean Squared Error).

***4. Entraînement du Modèle***

- Des images bruitées sont créées en ajoutant un bruit gaussien aux images d'entraînement.
- L'autoencodeur est ensuite entraîné sur ces images bruitées pour apprendre à les débruiter et reconstruire les images originales. Le modèle est entraîné pendant 10 époques, avec un batch size de 32.

***5. Débruitage des Images de Test***

- Le modèle entraîné est utilisé pour débruiter les images de test, qui ont également été bruitées avec du bruit gaussien.
- Le processus de prédiction génère des images débruitées qui sont comparées aux images originales pour évaluer la performance du modèle.

***6. Évaluation et Résultats***

- Une évaluation visuelle est réalisée en affichant des séries d'images : les images bruitées, les images débruitées par l'autoencodeur et les images originales.
- Les résultats sont présentés dans une grille où chaque ligne affiche respectivement l'image bruitée, l'image débruitée et l'image originale, permettant une comparaison claire de la qualité du débruitage.

***7. Optimisations Supplémentaires***

-
***Modularité :*** Le code est conçu de manière modulaire pour faciliter les ajustements futurs, comme l'ajout de nouvelles méthodes de traitement ou l'expérimentation avec d'autres architectures de modèles.
- 
***Amélioration de la gestion des erreurs :*** Le prétraitement des images a été amélioré pour gérer différents formats d'image (grayscale, RGBA) et garantir leur compatibilité avec le modèle.
- 
***Visualisation améliorée :*** La fonction de visualisation affiche les images de test sous forme de grille, ce qui améliore la présentation des résultats et permet une comparaison visuelle facile entre les images bruitées, débruitées et originales.

***8. Conclusion***

Le modèle d'autoencodeur convolutif a été efficace pour débruiter les images, réduisant le bruit de manière significative et reconstruisant des images proches des originales. Des optimisations supplémentaires peuvent être apportées pour améliorer encore la précision et la performance du modèle.

```python
import os
import imageio.v2 as imageio
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from scipy.ndimage import gaussian_filter
from skimage.restoration import denoise_nl_means, estimate_sigma
from skimage.util import random_noise
from scipy.ndimage import median_filter

# Fonction pour charger et prétraiter les images
def load_images_from_folder(folder, label=None, img_size=(128, 128)):
    images, labels, filenames = [], [], []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        if os.path.isfile(img_path):
            img = imageio.imread(img_path)

            if len(img.shape) == 2:
                img = np.stack((img,) * 3, axis=-1)
            elif img.shape[2] == 4:
                img = img[:, :, :3]

            img = tf.image.resize(img, img_size, method='bicubic') / 255.0
            images.append(img)
            filenames.append(filename)

            if label is not None:
                labels.append(label)

    return np.array(images), np.array(labels), filenames

folders = {
    "CATS": 0,
    "DOGS": 1,
    "PORTRAIT": 2,
    "CARS": 3,
    "FLOWERS": 4
}

base_folder = "C:/Users/ASUS/Downloads/A5/DATASETS"

images, labels = [], []
for folder_name, label in folders.items():
    folder_path = os.path.join(base_folder, folder_name)
    imgs, lbls, _ = load_images_from_folder(folder_path, label=label)
    images.extend(imgs)
    labels.extend(lbls)

images = np.array(images)
labels = np.array(labels)

X_train, X_val, y_train, y_val = train_test_split(images, labels, test_size=0.2, random_state=42)

folder_test = os.path.join(base_folder, "Images-test")
images_test, _, filenames_test = load_images_from_folder(folder_test, img_size=(128, 128))

augmentation = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
augmentation.fit(X_train)

# Définition de l'autoencodeur pour le débruitage
input_img = layers.Input(shape=(128, 128, 3))

# Encodeur
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = layers.MaxPooling2D((2, 2), padding='same')(x)
x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = layers.MaxPooling2D((2, 2), padding='same')(x)

# Décodeur
x = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(x)
x = layers.UpSampling2D((2, 2))(x)
x = layers.Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = layers.UpSampling2D((2, 2))(x)
decoded = layers.Conv2D(3, (3, 3), activation='sigmoid', padding='same')(x)

autoencoder = models.Model(input_img, decoded)

autoencoder.compile(optimizer='adam', loss='mse')

# Entraînement de l'autoencodeur sur des images bruitées
noisy_train = [random_noise(img, mode='gaussian', var=0.01) for img in X_train]
noisy_train = np.array(noisy_train)

autoencoder.fit(noisy_train, X_train, epochs=10, batch_size=32, shuffle=True, validation_split=0.1)

# Débruitage des images de test avec l'autoencodeur
noisy_test = [random_noise(img, mode='gaussian', var=0.01) for img in images_test[:8]]
noisy_test = np.array(noisy_test)
autoencoder_denoised = autoencoder.predict(noisy_test)

# Visualisation des résultats
fig, axes = plt.subplots(3, 8, figsize=(20, 8))
for i in range(8):
    axes[0, i].imshow(noisy_test[i])
    axes[0, i].set_title("Image Bruitée")
    axes[1, i].imshow(autoencoder_denoised[i])
    axes[1, i].set_title("Autoencodeur")
    axes[2, i].imshow(images_test[i])
    axes[2, i].set_title("Originale")
    for j in range(3):
        axes[j, i].axis('off')
plt.tight_layout()
plt.show()

print("Débruitage avec autoencodeur terminé.")
```

```python
import os
import imageio.v2 as imageio
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout

# 1. Charger et prétraiter les données
def load_images_from_folder(folder, label=None, img_size=(128, 128)):
    images = []
    labels = []
    filenames = []
    original_images = []  # Pour stocker les images brutes
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        if os.path.isfile(img_path):
            try:
                img = imageio.imread(img_path)  # Lire l'image
                original_images.append(img)  # Stocker l'image brute
                if len(img.shape) == 2:  # Si l'image est en niveaux de gris
                    img = np.stack((img,) * 3, axis=-1)  # Convertir en RGB
                elif img.shape[2] == 4:  # Si l'image a un canal alpha (RGBA)
                    img = img[:, :, :3]  # Garder seulement les 3 canaux RGB
                img = tf.image.resize(img, img_size, method='bicubic')  # Redimensionner
                img = preprocess_input(img)  # Normaliser les pixels pour ResNet50
                images.append(img)
                filenames.append(filename)
                if label is not None:  # Pour les dossiers catégorisés
                    labels.append(label)
            except Exception as e:
                print(f"Erreur lors du traitement de l'image {filename}: {e}")
    return images, labels, filenames, original_images

# Dossiers des catégories
folder_cats = "C:/Users/ASUS/Downloads/A5/DATASETS/CATS"
folder_dogs = "C:/Users/ASUS/Downloads/A5/DATASETS/DOGS"
folder_portraits = "C:/Users/ASUS/Downloads/A5/DATASETS/PORTRAIT"
folder_cars = "C:/Users/ASUS/Downloads/A5/DATASETS/CARS"
folder_flowers = "C:/Users/ASUS/Downloads/A5/DATASETS/FLOWERS"
folder_test = "C:/Users/ASUS/Downloads/A5/DATASETS/Images-test"

# Charger les données catégorisées (train/validation)
images_cats, labels_cats, _, _ = load_images_from_folder(folder_cats, label=0)
images_dogs, labels_dogs, _, _ = load_images_from_folder(folder_dogs, label=1)
images_portraits, labels_portraits, _, _ = load_images_from_folder(folder_portraits, label=2)
images_cars, labels_cars, _, _ = load_images_from_folder(folder_cars, label=3)
images_flowers, labels_flowers, _, _ = load_images_from_folder(folder_flowers, label=4)

# Convertir en arrays NumPy après avoir homogénisé les images
images = np.array(images_cats + images_dogs + images_portraits + images_cars + images_flowers)
labels = np.array(labels_cats + labels_dogs + labels_portraits + labels_cars + labels_flowers)

# Diviser les données en train/validation
X_train, X_val, y_train, y_val = train_test_split(images, labels, test_size=0.2, random_state=42)

# Charger les images du dossier Test (non catégorisées)
images_test, _, filenames_test, original_images_test = load_images_from_folder(folder_test, img_size=(128, 128))

# 2. Augmentation des données
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2,
    shear_range=0.2,
    brightness_range=(0.8, 1.2)
)
datagen.fit(X_train)

# 3. Construire le modèle CNN avec ResNet50 en transfert learning
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(128, 128, 3))
base_model.trainable = False  # Geler les couches de ResNet50

# Ajouter des couches personnalisées
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(5, activation='softmax')(x)  # 5 classes pour Cats, Dogs, Portraits, Cars, Flowers

model = Model(inputs=base_model.input, outputs=predictions)

# 4. Compiler et entraîner le modèle
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(datagen.flow(X_train, y_train, batch_size=32),
                    epochs=15,
                    validation_data=(X_val, y_val))

# 5. Évaluer le modèle
val_loss, val_accuracy = model.evaluate(X_val, y_val)
print(f"Validation Accuracy: {val_accuracy * 100:.2f}%")
# 6. Afficher les prédictions avec images originales
def display_test_predictions(original_images, filenames, model, img_size=(128, 128)):
    plt.figure(figsize=(12, 12))
    for i in range(min(9, len(original_images))):  # Afficher jusqu'à 9 images
        img = original_images[i]  # Utiliser l'image originale
        img_resized = tf.image.resize(img, img_size)  # Redimensionner uniquement pour la prédiction
        prediction = model.predict(np.expand_dims(img_resized, axis=0), verbose=0)
        
        # Récupérer la classe prédite avec la plus haute probabilité
        pred_label = np.argmax(prediction)
        
        # Afficher l'image et la prédiction
        plt.subplot(3, 3, i + 1)
        plt.imshow(img.astype('uint8'))  # Afficher l'image originale
        plt.title(f"{filenames[i]}\nPrédit: {['Cats', 'Dogs', 'Portraits', 'Cars', 'Flowers'][pred_label]}")
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Afficher les prédictions sur le dossier Test
display_test_predictions(original_images_test, filenames_test, model)

# Remarques :
# - Les erreurs du modèle pourraient être dues à :
#   - Une mauvaise qualité ou diversité des images.
#   - Des images similaires entre classes (par exemple, un portrait d'une personne ressemblant à une peinture).
# - Une inspection des échantillons mal classifiés pourrait révéler des schémas à corriger.

```
