## ***Introduction du Projet : Classification Automatique d'Images à l'Aide de CNN***

Avec l'évolution rapide des technologies numériques et l'augmentation exponentielle des données visuelles, la classification automatique d'images est devenue un domaine clé dans le champ de l'intelligence artificielle. Ce projet vise à concevoir et entraîner un modèle de réseau de neurones convolutifs (CNN) pour classifier automatiquement des images dans plusieurs catégories : chats, chiens, portraits, voitures, et fleurs.

Le défi principal de cette tâche réside dans la capacité du modèle à identifier des caractéristiques visuelles distinctives propres à chaque classe tout en étant robuste face à la variabilité des images (différents angles, luminosité, dimensions, etc.). Pour relever ce défi, nous utilisons un pipeline complet de deep learning, incluant le prétraitement des données, l'augmentation des images, et l'entraînement d'un modèle CNN optimisé.

## **Résumé des Étapes Réalisées dans le Projet : Classification Automatique d'Images**
Ce projet vise à développer une solution efficace pour la classification d'images en cinq catégories : chats, chiens, portraits, voitures, et fleurs, en utilisant des réseaux neuronaux convolutifs (CNN). Voici un résumé détaillé des étapes réalisées :

**1. Chargement et Prétraitement des Données**

- Chargement des images : Les images provenant de différents dossiers ont été chargées à l'aide d'une fonction dédiée. Chaque dossier correspond à une catégorie (chats, chiens, etc.).
- Homogénéisation des images :

  -  ***Redimensionnement*** : Toutes les images ont été redimensionnées à une taille uniforme de 128x128 pixels pour assurer une entrée cohérente dans le réseau.
  -  ***Normalisation*** : Les valeurs des pixels ont été ramenées entre 0 et 1 en divisant chaque pixel par 255. Cela aide à accélérer l'entraînement du modèle.
  -  ***Gestion des cas particuliers*** : Les images en niveaux de gris ont été converties en RGB en dupliquant les canaux, et les images avec un canal alpha (RGBA) ont été réduites à trois canaux RGB.

**2. Division des Données**

Les données ont été divisées en deux ensembles :
- ***Ensemble d'entraînement (80%)*** : Utilisé pour entraîner le modèle.
- ***Ensemble de validation (20%)*** : Utilisé pour évaluer les performances du modèle sur des données non vues pendant l'entraînement.

Une fonction de test a été mise en place pour charger des images supplémentaires destinées à tester le modèle.

***3. Augmentation des Données***

Pour augmenter la diversité des données et améliorer la robustesse du modèle, des transformations aléatoires ont été appliquées aux images d'entraînement à l'aide de la classe ImageDataGenerator :

- Rotation aléatoire jusqu'à 30 degrés.
- Translation horizontale et verticale jusqu'à 30% de l'image.
- Zoom aléatoire jusqu'à 20% de l'image.
- Effets de cisaillement (shear).
- Flip horizontal (miroir).

Ces techniques permettent d’éviter le sur-apprentissage (overfitting) en rendant le modèle plus généraliste.

***4. Construction du Modèle CNN***

Un modèle CNN a été conçu avec les éléments suivants :

- ***Convolutions et pooling :***
    - Trois blocs successifs de couches Conv2D avec des filtres croissants (32, 64, 128).
    - Chaque couche de convolution est suivie d'une couche MaxPooling2D pour réduire la dimension spatiale et conserver les caractéristiques importantes.

- ***Flatten et couches fully connected :***

    - Les sorties des convolutions ont été aplaties (Flatten).
    - Une couche dense avec 128 neurones et une activation ReLU a été ajoutée.
    - Une couche de Dropout (50%) a été utilisée pour réduire le sur-apprentissage.

- ***Sortie :***

    - Une couche dense finale avec 5 neurones (correspondant aux 5 catégories) et une activation softmax pour produire des probabilités.

***5. Entraînement et Optimisation***

- Le modèle a été compilé avec l'optimiseur Adam (learning rate de 0.0001), une fonction de perte sparse_categorical_crossentropy (adaptée aux labels entiers), et l'indicateur de performance accuracy.
- L'entraînement a été réalisé sur 20 époques en utilisant les données augmentées.
- Les courbes de perte et d’accuracy pour les ensembles d'entraînement et de validation ont été tracées pour analyser les performances.

***6. Évaluation et Résultats***

- Le modèle a été évalué sur l’ensemble de validation, atteignant une accuracy de validation satisfaisante.
- Une évaluation visuelle a été effectuée sur un ensemble d’images de test non catégorisées, avec une fonction qui affiche les images accompagnées des prédictions du modèle.


***7. Optimisations Supplémentaires***

- ***Simplification et modularité :*** Le code a été optimisé pour rendre chaque partie indépendante et modulaire.
- ***Gestion des erreurs :*** Le traitement des images (grayscale, RGBA) et des cas particuliers a été anticipé.
- ***Visualisation améliorée :*** Une fonction a été ajoutée pour afficher les images de test sous forme de grille avec leurs prédictions, garantissant une présentation claire et esthétique des résultats.

### **le code complet**

```python
import os
import imageio.v2 as imageio
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Fonction pour charger et prétraiter les images
def load_images_from_folder(folder, label=None, img_size=(128, 128)):
    images, labels, filenames = [], [], []
    for filename in os.listdir(folder):
        img_path = os.path.join(folder, filename)
        if os.path.isfile(img_path):
            img = imageio.imread(img_path)

            # Gérer les images en niveaux de gris ou avec un canal alpha
            if len(img.shape) == 2:
                img = np.stack((img,) * 3, axis=-1)
            elif img.shape[2] == 4:
                img = img[:, :, :3]

            # Redimensionner et normaliser
            img = tf.image.resize(img, img_size, method='bicubic') / 255.0
            images.append(img)
            filenames.append(filename)

            if label is not None:
                labels.append(label)

    return np.array(images), np.array(labels), filenames

# Dossiers des catégories
folders = {
    "CATS": 0,
    "DOGS": 1,
    "PORTRAIT": 2,
    "CARS": 3,
    "FLOWERS": 4
}
base_folder = "C:/Users/ASUS/Downloads/A5/DATASETS"

# Charger les images et labels
images, labels = [], []
for folder_name, label in folders.items():
    folder_path = os.path.join(base_folder, folder_name)
    imgs, lbls, _ = load_images_from_folder(folder_path, label=label)
    images.extend(imgs)
    labels.extend(lbls)

images = np.array(images)
labels = np.array(labels)

# Diviser les données en ensembles d'entraînement et de validation
X_train, X_val, y_train, y_val = train_test_split(images, labels, test_size=0.2, random_state=42)

# Charger les images de test
folder_test = os.path.join(base_folder, "Images-test")
images_test, _, filenames_test = load_images_from_folder(folder_test, img_size=(128, 128))

# Augmentation des données
augmentation = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
augmentation.fit(X_train)

# Construction du modèle CNN
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(len(folders), activation='softmax')
])

# Résumé du modèle
model.summary()

# Compilation du modèle
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Entraînement du modèle
history = model.fit(
    augmentation.flow(X_train, y_train, batch_size=32),
    epochs=20,
    validation_data=(X_val, y_val),
    verbose=1
)

# Évaluation du modèle
val_loss, val_accuracy = model.evaluate(X_val, y_val, verbose=0)
print(f"Validation Accuracy: {val_accuracy * 100:.2f}%")

# Tracer les courbes de perte et d'accuracy
plt.figure(figsize=(12, 6))

# Erreur
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Erreur Entra\u00eenement')
plt.plot(history.history['val_loss'], label='Erreur Validation')
plt.xlabel('Épochs')
plt.ylabel('Erreur')
plt.title("\u00c9volution de l'erreur")
plt.legend()

# Accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Accuracy Entra\u00eenement')
plt.plot(history.history['val_accuracy'], label='Accuracy Validation')
plt.xlabel('Épochs')
plt.ylabel('Accuracy')
plt.title("\u00c9volution de l'accuracy")
plt.legend()

plt.tight_layout()
plt.show()

# Fonction pour afficher les prédictions sur les images de test
def display_test_predictions(images, filenames, model, img_size=(128, 128), grid_size=(4, 4)):
    plt.figure(figsize=(12, 12))
    num_images = min(len(images), grid_size[0] * grid_size[1])

    for i in range(num_images):
        original_img = images[i]
        resized_img = tf.image.resize(original_img, img_size)
        prediction = model.predict(np.expand_dims(resized_img, axis=0), verbose=0)
        pred_label = np.argmax(prediction)

        plt.subplot(grid_size[0], grid_size[1], i + 1)
        plt.imshow(original_img)
        plt.title(f"{filenames[i]}\nPr\u00e9dit: {list(folders.keys())[pred_label]}")
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# Afficher les prédictions
print("Affichage des prédictions pour les images de test :")
display_test_predictions(images_test, filenames_test, model, grid_size=(4, 4))
```
